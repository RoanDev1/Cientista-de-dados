```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(3, 2)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=sinasc)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=sinasc)
sns.countplot(ax = axes[1, 0], x='SEXO', data=sinasc)
sns.countplot(ax = axes[1, 1], x='GESTACAO', data=sinasc)
sns.countplot(ax = axes[2, 1], x='ESTCIVMAE', data=sinasc)

```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
media = sinasc.groupby('CONSULTAS')['APGAR5'].mean().reset_index()
media

sns.barplot(x='CONSULTAS', y='APGAR5', data=media)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

sinasc['SEXO'].value_counts().plot(kind='bar', ax=axes[0, 0], color='purple')
axes[0, 0].set_title('Frequência de SEXO')
axes[0, 0].set_xlabel('Sexo')
axes[0, 0].set_ylabel('Frequência')

sinasc['GESTACAO'].value_counts().plot(kind='bar', ax=axes[0, 1], color='yellow')
axes[0, 1].set_title('Frequência de GESTAÇÃO')
axes[0, 1].set_xlabel('Gestação')
axes[0, 1].set_ylabel('Frequência')

sinasc['ESTCIVMAE'].value_counts().plot(kind='bar', ax=axes[1, 0], color='blue')
axes[1, 0].set_title('Frequência de ESTCIVAMAE')
axes[1, 0].set_xlabel('Estado Civil')
axes[1, 0].set_ylabel('Frequência')

sinasc['CONSULTAS'].value_counts().plot(kind='bar', ax=axes[1, 1])
axes[1, 1].set_title('Frequência de CONSULTAS')
axes[1, 1].set_xlabel('Número de Consultas')
axes[1, 1].set_ylabel('Frequência')

plt.show
```
