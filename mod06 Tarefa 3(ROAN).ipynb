#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# %matplotlib notebook

sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sinasc.loc[(sinasc['APGAR5'] >=0) & (sinasc['APGAR5'] <= 3), 'CATEGORIA_ASFIXIA'] = 'Asfixia Grave'
sinasc.loc[(sinasc['APGAR5'] >=4) & (sinasc['APGAR5'] <= 5), 'CATEGORIA_ASFIXIA'] = 'Asfixia Moderada'
sinasc.loc[(sinasc['APGAR5'] >=6) & (sinasc['APGAR5'] <= 7), 'CATEGORIA_ASFIXIA'] = 'Asfixia Leve'
sinasc.loc[(sinasc['APGAR5'] >=8) & (sinasc['APGAR5'] <= 10), 'CATEGORIA_ASFIXIA'] = 'Asfixia Normal'
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
catplot = sns.catplot(x='GESTACAO', y='APGAR5'
            , hue = 'SEXO'
            , kind='bar'
            , estimator = sum
            , data=sinasc
           )

catplot.set_xticklabels(rotation=45)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tabela_cruzada = pd.crosstab(index=sinasc["GESTACAO"], columns=sinasc["APGAR5"])
print(tabela_cruzada)
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
freq_pctlin = tabela_cruzada.div(tabela_cruzada.sum(axis = 1), axis = 0)
print(freq_pctlin)

freq_pctlin.plot.bar()
```

```python
freq_pctlin.plot.bar()

# De 32 a 36 semanas, o APGAR5 dos bebês, em média, é bom, sendo a maoiria em 9.0. Porém, em gestações mais curtas, vemos que a variaçãoé maior, apesar de ainda ser positivo em sua maioria.
```
